Write a python program to collect text data from **either of the following sources** and save the data into a **csv file:**

(1) Collect all the customer reviews of a product (you can choose any porduct) on amazon. [atleast 1000 reviews]

(2) Collect the top 1000 User Reviews of a movie recently in 2024 or 2025 (you can choose any movie) from IMDB. [If one movie doesn't have sufficient reviews, collect reviews of atleast 2 or 3 movies]


(3) Collect the **abstracts** of the top 10000 research papers by using the query "machine learning", "data science", "artifical intelligence", or "information extraction" from Semantic Scholar.

```python
# Taks 1.2 Collect the top 1000 User Reviews of a movie recently in 2024 or 2025 (you can choose any movie) from IMDB. [If one movie doesn't have sufficient reviews, collect reviews of atleast 2 or 3 movies]

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time
import pandas as pd

# Movie URL (Dune: Part Two has 1000+ reviews)
url = "https://www.imdb.com/title/tt6263850/reviews"

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get(url)

time.sleep(5)

reviews = []

while len(reviews) < 1000:
    try:
        load_more = driver.find_element(By.CLASS_NAME, "ipc-see-more__button")
        driver.execute_script("arguments[0].click();", load_more)
        time.sleep(2)
    except:
        break

    soup = BeautifulSoup(driver.page_source, "html.parser")
    review_blocks = soup.find_all("div", class_="ipc-html-content-inner-div")

    reviews = [r.get_text(strip=True) for r in review_blocks]

    print("Collected:", len(reviews))

driver.quit()

reviews = reviews[:1000]

df = pd.DataFrame({"review": reviews})
df.to_csv("imdb_1000_reviews.csv", index=False)

print("Final Count:", len(reviews))
```

# Question 2 (15 points)

Write a python program to **clean the text data** you collected in the previous question and save the clean data in a new column in the csv file. The data cleaning steps include: [Code and output is required for each part]

(1) Remove noise, such as special characters and punctuations.

(2) Remove numbers.

(3) Remove stopwords by using the stopwords list.

(4) Lowercase all texts

(5) Stemming.

(6) Lemmatization.

```python
# Write code for each of the sub parts with proper comments.
import pandas as pd
import re
import nltk

from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

# Load Dataset
df = pd.read_csv("imdb_1000_reviews.csv")

# Display the first 5 rows of the DataFrame
df.head()
```

```python
# Task 2.1 Remove noise, such as special characters and punctuations

def remove_noise(text):
    return re.sub(r'[^\w\s]', '', str(text))

df["clean_step1"] = df["review"].apply(remove_noise)

df[["review", "clean_step1"]].head()
```

```python
# Task 2.2 Remove numbers

def remove_numbers(text):
    return re.sub(r'\d+', '', text)

df["clean_step2"] = df["clean_step1"].apply(remove_numbers)

df[["clean_step1", "clean_step2"]].head()
```

```python
# Task 2.3 Remove stopwords by using the stopwords list

stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    tokens = word_tokenize(text)
    filtered = [word for word in tokens if word.lower() not in stop_words]
    return " ".join(filtered)

df["clean_step3"] = df["clean_step2"].apply(remove_stopwords)

df[["clean_step2", "clean_step3"]].head()
```

```python
# Task 2.4 Lowercase all texts
df["clean_step4"] = df["clean_step3"].str.lower()

df[["clean_step3", "clean_step4"]].head()
```

```python
# Task 2.5 Stemming

stemmer = PorterStemmer()

def apply_stemming(text):
    tokens = word_tokenize(text)
    stemmed = [stemmer.stem(word) for word in tokens]
    return " ".join(stemmed)

df["clean_step5"] = df["clean_step4"].apply(apply_stemming)

df[["clean_step4", "clean_step5"]].head()
```

```python
# Task 2.6 Lemmatization

lemmatizer = WordNetLemmatizer()

def apply_lemmatization(text):
    tokens = word_tokenize(text)
    lemmatized = [lemmatizer.lemmatize(word) for word in tokens]
    return " ".join(lemmatized)

df["clean_text"] = df["clean_step4"].apply(apply_lemmatization)

df[["clean_step4", "clean_text"]].head()
```

```python
# Save Final Cleaned CSV

df.to_csv("imdb_1000_reviews_cleaned.csv", index=False)

print("Cleaned file saved successfully!")
```

# Question 3 (15 points)

Write a python program to **conduct syntax and structure analysis of the clean text** you just saved above. The syntax and structure analysis includes:

(1) **Parts of Speech (POS) Tagging:** Tag Parts of Speech of each word in the text, and calculate the total number of N(oun), V(erb), Adj(ective), Adv(erb), respectively.

(2) **Constituency Parsing and Dependency Parsing:** print out the constituency parsing trees and dependency parsing trees of all the sentences. Using one sentence as an example to explain your understanding about the constituency parsing tree and dependency parsing tree.

(3) **Named Entity Recognition:** Extract all the entities such as person names, organizations, locations, product names, and date from the clean texts, calculate the count of each entity.

```python
# Your code here
import nltk
from nltk import pos_tag
import pandas as pd
from nltk.tokenize import word_tokenize

df = pd.read_csv("imdb_1000_reviews_cleaned.csv")
df.head()
```

```python
# Task 3.1 Apply POS Tagging

def pos_tag_text(text):
    tokens = word_tokenize(str(text))
    return pos_tag(tokens)

df["pos_tags"] = df["clean_text"].apply(pos_tag_text)

df[["clean_text", "pos_tags"]].head()
```

```python
# Count Nouns, Verbs, Adjectives, Adverbs
def count_pos(tags):
    noun = 0
    verb = 0
    adj = 0
    adv = 0

    for word, tag in tags:
        if tag.startswith('NN'):
            noun += 1
        elif tag.startswith('VB'):
            verb += 1
        elif tag.startswith('JJ'):
            adj += 1
        elif tag.startswith('RB'):
            adv += 1

    return pd.Series([noun, verb, adj, adv])

df[["Noun_Count", "Verb_Count", "Adj_Count", "Adv_Count"]] = df["pos_tags"].apply(count_pos)

# Calculate Total Counts
total_nouns = df["Noun_Count"].sum()
total_verbs = df["Verb_Count"].sum()
total_adj = df["Adj_Count"].sum()
total_adv = df["Adv_Count"].sum()

print("Total Nouns:", total_nouns)
print("Total Verbs:", total_verbs)
print("Total Adjectives:", total_adj)
print("Total Adverbs:", total_adv)
```

```python
# Task 3,2 Constituency Parsing and Dependency Parsing
# The original review text was used for syntax analysis because heavy text cleaning may distort sentence structure and negatively affect constituency and dependency parsing. Therefore, the first review was selected for analysis.

# Constituency Parsing using Benepar
import warnings
warnings.filterwarnings("ignore")

import spacy
import benepar

# Load spaCy English model
nlp = spacy.load("en_core_web_sm")

# Add Benepar constituency parser
nlp.add_pipe("benepar", config={"model": "benepar_en3"})

# Select the first review (original text)
text = df["review"].iloc[0]

# Process text
doc = nlp(text)

# Print constituency tree for each sentence
for i, sent in enumerate(doc.sents):
    print("\n==============================")
    print(f"Sentence {i+1}: {sent.text}")
    print("==============================")
    print(sent._.parse_string)
```

#### We use the first sentence of this review as an example:“Hugh Jackman is the perfect Wolverine.”
#### The constituency parsing tree shows the phrase structure of the sentence. The whole sentence is labeled as S. It is divided into a noun phrase (NP), “Hugh Jackman,” which functions as the subject, and a verb phrase (VP), “is the perfect Wolverine.” Within the VP, “the perfect Wolverine” forms another noun phrase. This analysis illustrates how words are grouped into hierarchical phrase structures within a sentence

```python
# Dependency Parsing using spaCy
# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Select the first review
text = df["review"].iloc[0]

# Process text
doc = nlp(text)

# Print dependency relations for each sentence
for i, sent in enumerate(doc.sents):
    print("\n==============================")
    print(f"Sentence {i+1}: {sent.text}")
    print("==============================")

    for token in sent:
        print(token.text, "-->", token.dep_, "-->", token.head.text)
```

#### We use the same sentence as an example: “Hugh Jackman is the perfect Wolverine.”

#### The dependency parsing tree shows the grammatical relationships between words. The main verb “is” is labeled as the ROOT, meaning it is the central word of the sentence. “Jackman” is the subject (nsubj) of “is,” and “Hugh” modifies “Jackman” as a compound name. “Wolverine” functions as the complement (attr) of the verb “is.” The word “perfect” modifies “Wolverine” as an adjective (amod), and “the” acts as a determiner (det). This analysis focuses on how words are connected to each other and their grammatical roles in the sentence.

```python
# Task3.3 Named Entity Recognition
# To better identify and extract entities such as person names, organizations, locations, product names, and dates, we chose to perform the analysis on the original review text. This ensures that the full contextual and structural information of the sentences is preserved, allowing for more accurate named entity recognition.

# Load the English language model
nlp = spacy.load("en_core_web_sm")

# Extract Named Entities
def extract_entities(text):
    doc = nlp(str(text))
    entities = []

    for ent in doc.ents:
        entities.append((ent.text, ent.label_))

    return entities

# Apply entity extraction to original reviews
df["entities"] = df["review"].apply(extract_entities)

df[["review", "entities"]].head()
```

```python
from collections import Counter

# Flatten all entities into a single list
all_entities = []

for entity_list in df["entities"]:
    all_entities.extend(entity_list)

# Count entities by label
entity_counts = Counter([label for text, label in all_entities])

print("Entity Counts by Type:")
print(entity_counts)
```

Q4. (PART-1)
Web scraping data from the GitHub Marketplace to gather details about popular actions. Using Python, the process begins by sending HTTP requests to multiple pages of the marketplace (1000 products), handling pagination through dynamic page numbers. The key details extracted include the product name, a short description, and the URL.

 The extracted data is stored in a structured CSV format with columns for product name, description, URL, and page number. A time delay is introduced between requests to avoid server overload. ChatGPT can assist by helping with the parsing of HTML, error handling, and generating reports based on the data collected.

 The goal is to complete the scraping within a specified time limit, ensuring that the process is efficient and adheres to GitHub’s usage guidelines.

(PART -2)

1.   **Preprocess Data**: Clean the text by tokenizing, removing stopwords, and converting to lowercase.

2. Perform **Data Quality** operations.


Preprocessing:
Preprocessing involves cleaning the text by removing noise such as special characters, HTML tags, and unnecessary whitespace. It also includes tasks like tokenization, stopword removal, and lemmatization to standardize the text for analysis.

Data quality checks ensure completeness, consistency, and accuracy by verifying that all required columns are filled and formatted correctly. Additionally, it involves identifying and removing duplicates, handling missing values, and ensuring the data reflects the true content accurately.


Github MarketPlace page:
https://github.com/marketplace?type=actions

```python
import time
import random
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.common.exceptions import NoSuchElementException

# Configure Chrome WebDriver
chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")

# Initialize WebDriver using ChromeDriverManager
driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install()),
    options=chrome_options
)

# Base URL with dynamic pagination
BASE_URL = "https://github.com/marketplace?type=actions&page={}"

# Store all scraped data
all_data = []

# Set number of pages to scrape
# Adjust this carefully to avoid overloading the server
max_pages = 55

# Start Scraping Process
for page in range(1, max_pages + 1):
    url = BASE_URL.format(page)
    if page % 10 == 0:
        print(f"Scraping page {page}...")

    try:
        # Load the page
        driver.get(url)

        # Wait for dynamic JavaScript content to load
        time.sleep(random.uniform(3, 5))

        # Locate all product cards
        # Each card contains title + description
        items = driver.find_elements(
            By.CSS_SELECTOR,
            "div[data-testid='non-featured-item']"
        )

        # Stop if no products found
        if not items:
            print("No more products found.")
            break

        # Loop through each product card
        for item in items:
            try:
                # Extract product name
                name = item.find_element(
                    By.CSS_SELECTOR,
                    "a.line-clamp-1"
                ).text.strip()

                # Extract description
                try:
                    description = item.find_element(
                        By.CSS_SELECTOR,
                        "p.line-clamp-2"
                    ).text.strip()
                except NoSuchElementException:
                    description = "N/A"

                # Extract product URL
                link = item.find_element(
                    By.CSS_SELECTOR,
                    "a.line-clamp-1"
                ).get_attribute("href")

                # Append to list
                all_data.append({
                    "product_name": name,
                    "description": description,
                    "url": link,
                    "page_number": page
                })

            except Exception as e:
                print(f"Error extracting product: {e}")
                continue

    except Exception as e:
        print(f"Error loading page {page}: {e}")
        continue

    # Introduce delay between page requests
    time.sleep(random.uniform(2, 4))


# Close Browser
driver.quit()

# Convert data to DataFrame
df = pd.DataFrame(all_data)

# Save data to CSV
df.to_csv("github_marketplace_actions.csv", index=False)

print("\nScraping completed.")
print(f"Total records collected: {len(df)}")
```

```python
# Task 4.1 Preprocess Data

import re
import pandas as pd
from bs4 import BeautifulSoup
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Initialize tools
stop_words = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    if pd.isna(text):
        return ""

    # 1. Remove HTML tags
    text = BeautifulSoup(text, "html.parser").get_text()

    # 2. Remove special characters and numbers
    text = re.sub(r"[^a-zA-Z\s]", "", text)

    # 3. Convert to lowercase
    text = text.lower()

    # 4. Remove extra whitespace
    text = re.sub(r"\s+", " ", text).strip()

    # 5. Tokenization
    tokens = word_tokenize(text)

    # 6. Remove stopwords
    tokens = [word for word in tokens if word not in stop_words]

    # 7. Lemmatization
    tokens = [lemmatizer.lemmatize(word) for word in tokens]

    return " ".join(tokens)

# Apply preprocessing to description column
df["clean_description"] = df["description"].apply(preprocess_text)

# Save cleaned data
df.to_csv("github_actions_cleaned.csv", index=False)

print("Preprocessing completed successfully.")
```

```python
# Task 5.1
import tweepy
import pandas as pd
from dotenv import load_dotenv
import os

# 1. Authentication
# Load environment variables from a .env file
load_dotenv()

# Initialize Twitter API client with Bearer Token for authentication
client = tweepy.Client(bearer_token=os.getenv("TWITTER_X_BEARER_TOKEN"))

# 2. Define Search Query
# Search for hashtags related to Machine Learning or AI
query = "(#MachineLearning OR #ArtificialIntelligence) -is:retweet lang:en"

# 3. Fetch Tweets
tweets = client.search_recent_tweets(
    query=query,
    max_results=100,
    tweet_fields=["id", "text", "author_id"],
    expansions=["author_id"],
    user_fields=["username"]
)

users = {u["id"]: u["username"] for u in tweets.includes["users"]}

# 4. Store Data
data = []

if tweets.data:
    for tweet in tweets.data:
        data.append([
            tweet.id,
            users[tweet.author_id],
            tweet.text
        ])

# Create DataFrame
df = pd.DataFrame(data, columns=["tweet_id", "username", "text"])

# Save to CSV
df.to_csv("ml_ai_tweets.csv", index=False)

print("Tweet scraping completed successfully.")
print("Total tweets collected:", len(df))
```
